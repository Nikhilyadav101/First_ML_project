```
#importing the useful libraries 
import numpy as np 
import pandas as pd

```

```
#importing the dataset from sklearn library
from sklearn.datasets import load_iris

```

```
iris_dataset=load_iris()
```

```
print("the keys for the datasets are: {}".format(iris_dataset.keys()))
```

```
#to know the description of this dataset
print(iris_dataset['DESCR'])
```

```
print(iris_dataset['target_names'])
print("Data:{}".format(iris_dataset['data']))
print("Target:{}".format(iris_dataset['target']))
```

```
#spliting the datasets in train and test splits
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(iris_dataset['data'],iris_dataset['target'],random_state=0)
```

```
print("Shape of X_train:{}".format(X_train.shape))
print("Shape of y_train:{}".format(y_train.shape))
```

```
print("Shape of X_test:{}".format(X_test.shape))
print("Shape of y_test:{}".format(y_test.shape))
```

```
#create dataframe from data in X_train
#label the columns using the strings in iris_dataset.features_names
df=pd.DataFrame(X_train,columns=iris_dataset.feature_names)
```

```
#creating a scatter matrix from the dataframe,color by y_train
#pairplot of iris dataset
grr=pd.scatter_matrix(df,c=y_train, figsize=(15,15),marker='o',hist_kwds={'bins':20},s=60,alpha=.8)
```

```
#using the k-Nearest 
from sklearn.neighbors import KNeighborsClassifier
```

```
knn =KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train,y_train)
```

```
KNeighborsClassifier(algorithm='auto',leaf_size=30,metric='minkowski',metric_params=None,n_jobs=1,n_neighbors=1,p=2,weights='uniform')
```

```
X_new=np.array([[5,2.9,1,0.2]])
print("X_new.shape:{}".format(X_new.shape))
```

```
prediction=knn.predict(X_new)
print("Pridiction:{}".format(prediction))
print("Pridiction:{}".format(iris_dataset['target_names'][prediction]))
```

```
y_pred=knn.predict(X_test)
```

```
print("The predicted values are:{}".format(y_pred))
```

```
#now to check the accuracy 
print("Test set score:{}".format(np.mean(y_pred==y_test)))
```

```
#using knn.score method
print("Test score:{}".format(knn.score(X_test,y_test)))
```
